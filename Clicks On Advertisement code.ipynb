# Logistic Regression Project 

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

## Exploratory Data Analysis

Let's use seaborn to explore the data!

**Creating a histogram of the Age**

```python
sns.set_style("whitegrid")
sns.histplot(x=ad_data["Age"],bins=30)
```

**Creating a jointplot showing Area Income versus Age.**

```python
sns.jointplot(x="Age",y="Area Income",data=ad_data)
import scipy.stats as stats
r,p=stats.pearsonr(ad_data["Age"],ad_data["Area Income"])
plt.legend(["r={:.3f} , p={:.5f}".format(r,p)])
```

**Creating a jointplot showing the kde distributions of Daily Time spent on site vs. Age.**

```python
sns.jointplot(x="Age",y="Daily Time Spent on Site",data=ad_data,kind="kde",
              fill=True,color="red",marginal_kws={"lw":2})
r,p=stats.pearsonr(ad_data["Age"],ad_data["Daily Time Spent on Site"])
plt.legend(["r={:.3f} , p={:.5f}".format(r,p)])
```

**Creating a jointplot of 'Daily Time Spent on Site' vs. 'Daily Internet Usage**

```python
sns.jointplot(data=ad_data,x="Daily Time Spent on Site",
              y="Daily Internet Usage",color="green",marginal_kws={"alpha":0.5})
r,p=stats.pearsonr(ad_data["Daily Time Spent on Site"],
                   ad_data["Daily Internet Usage"])
plt.legend(["r={:.2f} , p={:.2}".format(r,p)])
```

**Finally, we create a pairplot with the hue defined by the 'Clicked on Ad' column feature.**

```python
plt.figure(figsize=(12,7))
sns.set_theme(style="darkgrid",context="notebook")
sns.pairplot(data=ad_data,hue="Clicked on Ad",diag_kind="hist")
plt.tight_layout()
```

# Logistic Regression

Now it's time to do a train test split, and train our model!

We'll have the freedom here to choose columns that we want to train on!

**Split the data into training set and testing set using train_test_split**

```python
ad_data.head()
X = ad_data.drop(["Timestamp","Ad Topic Line","City","Country","Clicked on Ad"],axis=1)
y = ad_data["Clicked on Ad"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

**Training and fiting a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=100,solver="lbfgs",verbose=100)
```

```python
log.fit(X_train,y_train)
```

## Predictions and Evaluations
**Now we predict values for the testing data.**

```python
predicted = log.predict(X_test)
```

**Creating a classification report for the model.**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predicted))
```

```python
print(confusion_matrix(y_test,predicted))
```
